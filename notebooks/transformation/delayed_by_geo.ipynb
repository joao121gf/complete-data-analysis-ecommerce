## ⚙️ Instalacoes de depêndencias e leitura de arquivos

```python
import pandas as pd
from IPython.display import display


path_customers = 'C:/Users/Gioher/Desktop/Programação/Projetos/complete-data-analysis-ecommerce/data/filtred/customers_filtred.csv'
path_orders = 'C:/Users/Gioher/Desktop/Programação/Projetos/complete-data-analysis-ecommerce/data/filtred/orders_filtred.csv'
delayed_orders = 'C:/Users/Gioher/Desktop/Programação/Projetos/complete-data-analysis-ecommerce/data/metrics/delayed_orders.csv'


customers_filtred = pd.read_csv(path_customers, sep=',')
orders_filtred = pd.read_csv(path_orders, sep=',')
delayed_orders = pd.read_csv(delayed_orders, sep=',')

```

```python
display(customers_filtred.head(2))
display(orders_filtred.head(2))
```

# ⚙️ Merge nos dois dataframes para extrair a cidade da qual foi realizada a ordem de compra
 

```python
merged_datas = pd.merge(customers_filtred, orders_filtred, on='customer_id', how='inner')
```

```python
display(merged_datas.head(2))
```

```python
# Filtrando apenas linhas com siglas de estado com exatamente 2 letras
merged_datas = merged_datas[merged_datas['customer_state'].str.len() == 2]
```

```python
#Tabela com pedidos atrasados sendo 7827 no total
display(delayed_orders.shape)
```

```python
display(delayed_orders.head(3))
```

```python
#Filtrando o merged_datas para obter apenas os pedidos atrasados
filtred_delayed = merged_datas[merged_datas['order_id'].isin(delayed_orders['order_id'])]
```

```python
display(filtred_delayed.head(4))
```

```python
#Filtrando pedidos por estado
orders_by_state = merged_datas.groupby('customer_state')['order_id'].count().sort_values(ascending=False)

display(orders_by_state.head(4))
```

```python
#Filtrando pedidos atrasados por estado
delayed_orders_by_state = filtred_delayed.groupby('customer_state')['order_id'].count().sort_values(ascending=False)
display(delayed_orders_by_state.head(10))
```

```python
df_percentual = pd.DataFrame({
    'total_orders': orders_by_state,
    'delayed_orders': delayed_orders_by_state
})
df_percentual = df_percentual[df_percentual.index.str.len() == 2]
display(df_percentual)
```

```python
# Preenchendo os NaNs com 0 caso alguma cidade não tenha pedidos atrasados
df_percentual['delayed_orders'] = df_percentual['delayed_orders'].fillna(0)
display(df_percentual)
```

```python
#Calculando a porcentagem de atrasados
df_percentual['delayed_percent'] = (df_percentual['delayed_orders'] / df_percentual['total_orders']) * 100

display(df_percentual)
```

```python
# Ordenando pela maior porcentagem de atrasos
percentual_delayed_by_state = df_percentual.sort_values(by='delayed_percent', ascending=False)
```

```python
display(percentual_delayed_by_state.head(10))
```

```python
#Exportando csv de df_percentual

percentual_delayed_by_state['delayed_percent'] = percentual_delayed_by_state['delayed_percent'] / 100


percentual_delayed_by_state['delayed_percent'] = percentual_delayed_by_state['delayed_percent'].round(4)


percentual_delayed_by_state.to_csv('percentual_delayed_by_state.csv', sep=',', decimal=',')

```
